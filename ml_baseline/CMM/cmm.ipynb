# Gausian Mixture Models Algorithm


```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np 
from sklearn import mixture
from sklearn.preprocessing import StandardScaler 
import sys
sys.path.append('../')
from shared_modules.gmm_graph_utils import plot_gmm, find_num_components
```

```python
import pandas as pd

weather_df = pd.read_csv('a1.txt', delim_whitespace = True)
#weather_df.iloc[:,0]
##weather_df
mat = StandardScaler().fit_transform(weather_df);
plt.scatter(mat[:,0],mat[:,1])
```

```python
# Find the number of components to pass to the GausianMixture module
find_num_components(mat, 35)
```

# The Bandwidth parameter is the most important to this algorithm.  It is a "window" that is described in many articles.  It changes for every dataset

```python
from sklearn.mixture import GaussianMixture
gmm = GaussianMixture(n_components=22).fit(mat)
labels = gmm.predict(mat)
plt.scatter(mat[:, 0], mat[:, 1], c=labels, s=40, cmap='viridis')
```

```python
probs = gmm.predict_proba(mat)
print(probs[:5].round(3))
```

```python
plt.title("Gaussian Mixture Algorithm")

gmm= GaussianMixture(n_components=22, covariance_type='full', random_state=42)
plot_gmm(gmm, mat)
```
